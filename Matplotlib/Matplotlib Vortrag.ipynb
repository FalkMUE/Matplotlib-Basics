# Einfache Plots

```python
y = [1,5,4,1,9]
plt.plot(y)
plt.show()
```

```python
y = [1,5,4,1,9]
plt.plot(y, color = 'red', marker = 'D', linestyle = '') #roten Punkte
plt.plot(y, color = 'green', linestyle = '--') #Verbindung 
plt.show()
```

```python
x = [3, 4, 7, 8, 10]
y = [5, 7, 7, 7, 11]
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 8, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x Achse')
plt.ylabel('y Achse')
plt.title("Einacher Plot")

plt.legend()
plt.axis([0, 9, 0, 1000]) #plt.axis([xmin, xmax, ymin, ymax])
plt.grid(True)
plt.show()
```

* Syntax von linspace :
* linspace(start, stop, num=50, endpoint=True, retstep=False)
* Syntax von axis:
* plt.axis([xmin, xmax, ymin, ymax])

# The Iris Dataset

* Diese Datensätze bestehen aus 3 verschiedenen Arten von Iris (Setosa, Versicolour und Virginica).
* Die Spalten sind: Sepal Length, Sepal Width, Petal Length and Petal Width.

```python
pandas.read_csv('../Matplotlib/iris.csv', delimiter = ',')  #Komma, da es eine CSV Datei ist !
```

```python
iris = pandas.read_csv('../Matplotlib/iris.csv', delimiter = ',')
```

* Iris Data set ist jetzt in der varaible iris gespeichert

```python
iris.info()
```

```python
print(iris.groupby('species').size()) 
```

* Jede Spezies hat 50 Reihen 

```python
setosa = iris[iris['species']=='setosa']
versicolor = iris[iris['species']=='versicolor']
virginica = iris[iris['species']=='virginica']

#print(setosa.describe())
#print(versicolor.describe())
#print(virginica.describe())
```

* DataFrame erzeugen für jede Spezies

```python
import matplotlib.image as mpimg

img = mpimg.imread('matplotlib-diagramm-aufbau.png')
imgplot = plt.imshow(img, aspect = 'auto')

plt.show()
```

```python
plt.figure()
fig,ax = plt.subplots(1,2,figsize=(17, 5))

iris.plot(x="sepal_length",y="sepal_width",kind="scatter",ax=ax[0],label="sepal",color='r')
iris.plot(x="petal_length",y="petal_width",kind="scatter",ax=ax[1],label="petal",color='b')



ax[0].set(title='Sepal Vergleich ')
ax[1].set(title='Petal Vergleich')
ax[0].legend()
ax[1].legend()
plt.show()
```

```python
plt.figure()
fig,ax = plt.subplots(1, 2, figsize = (17, 10))

setosa.plot(x="sepal_length", y="sepal_width", kind="scatter",ax=ax[0],label='setosa',color='r')
versicolor.plot(x="sepal_length",y="sepal_width",kind="scatter",ax=ax[0],label='versicolor',color='b')
virginica.plot(x="sepal_length", y="sepal_width", kind="scatter", ax=ax[0], label='virginica', color='g')

setosa.plot(x="petal_length", y="petal_width", kind="scatter",ax=ax[1],label='setosa',color='r')
versicolor.plot(x="petal_length",y="petal_width",kind="scatter",ax=ax[1],label='versicolor',color='b')
virginica.plot(x="petal_length", y="petal_width", kind="scatter", ax=ax[1], label='virginica', color='g')

ax[0].set(title='Sepal Vergleich ')
ax[1].set(title='Petal Vergleich')
ax[0].legend()
ax[1].legend()
plt.show()
```
